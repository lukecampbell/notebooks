```python
main_dir = '/Users/luke/Documents/Dev/code/glos/data/mtri_ranger3'

import os
from netCDF4 import Dataset

def generate_cdl(path):
    out = path.replace('.nc', '.cdl')
    from sh import ncgen
    ncgen('-o', out, path)
```

```python
nc = Dataset(os.path.join(main_dir, 'ranger3_2013.nc'), 'r+')
```

```python
nc.createDimension('points')
time_var = nc.createVariable('time', 'float', ('points',), fill_value=3.402823e+38)
time_var.units = 'seconds since 2000-01-01'
time_var.long_name = 'time of data data collection'
time_var.standard_name = 'time'
time_var.calendar = 'gregorian'

speed_var = nc.createVariable('speed', 'float', ('points',), fill_value=3.402823e+38)
speed_var.units = 'm/s'
speed_var.long_name = 'speed of the Ranger III'
speed_var.standard_name = 'speed'
speed_var.coordinates = 'time latitude longitude'

heading = nc.createVariable('heading', 'float', ('points',), fill_value=3.402823e+38)
heading.units = "Degrees" 
heading.long_name = "heading of the Ranger IIII" 
heading.standard_name = "heading" 
heading.coordinates = "time latitude longitude" 

air_temperature = nc.createVariable('air_temperature', 'float', ('points',), fill_value=3.402823e+38)
air_temperature.units = "Celsius" 
air_temperature.long_name = "temperature of the air" 
air_temperature.standard_name = "air_temperature" 
air_temperature.coordinates = "time latitude longitude" 

humidity = nc.createVariable('humidity', 'float', ('points',), fill_value=3.402823e+38)
humidity.units = "Percent" 
humidity.long_name = "relative humidity" 
humidity.standard_name = "humidity" 
humidity.coordinates = "time latitude longitude" 

pressure = nc.createVariable('pressure', 'float', ('points',), fill_value=3.402823e+38)
pressure.units = "hPa" 
pressure.long_name = "atmospheric pressure" 
pressure.standard_name = "pressure" 
pressure.coordinates = "time latitude longitude" 

wind_speed = nc.createVariable('wind_speed', 'float', ('points',), fill_value=3.402823e+38)
wind_speed.units = "m/s" 
wind_speed.long_name = "speed of the wind" 
wind_speed.standard_name = "wind_speed" 
wind_speed.coordinates = "time latitude longitude" 

wind_direction = nc.createVariable('wind_direction', 'float', ('points',), fill_value=3.402823e+38)
wind_direction.units = "Degrees" 
wind_direction.long_name = "heading of the wind" 
wind_direction.standard_name = "wind_direction" 
wind_direction.coordinates = "time latitude longitude" 

rainfall_intensity = nc.createVariable('rainfall_intensity', 'float', ('points',), fill_value=3.402823e+38)
rainfall_intensity.units = "mm per Hour" 
rainfall_intensity.long_name = "rainfall intensity" 
rainfall_intensity.standard_name = "rainfall_intensity" 
rainfall_intensity.coordinates = "time latitude longitude" 

accumulated_rainfall = nc.createVariable('accumulated_rainfall', 'float', ('points',), fill_value=3.402823e+38)
accumulated_rainfall.units = "mm" 
accumulated_rainfall.long_name = "accumulated rain" 
accumulated_rainfall.standard_name = "accumulated_rainfall" 
accumulated_rainfall.coordinates = "time latitude longitude" 

longitude = nc.createVariable('longitude', 'float', ('points',), fill_value=3.402823e+38)
longitude.units = "degrees_east" 
longitude.long_name = "Longitude" 
longitude.standard_name = "longitude" 

latitude = nc.createVariable('latitude', 'float', ('points',), fill_value=3.402823e+38)
latitude.units = "degrees_north" 
latitude.long_name = "latitude" 
latitude.standard_name = "latitude" 

```

```python
from csv import DictReader
from dateutil.parser import parse
from datetime import datetime
import re
csv_file = os.path.join(main_dir, 'ranger3_2013.csv')
f = open(csv_file, 'r')
dr = DictReader(f)


mapping = {
        'time' : 'weather_sensor_date_time',
        'speed': 'speed',
        'heading' : 'heading',
        'air_temperature' : 'air_temperature',
        'humidity' : 'humidity',
        'pressure' : 'pressure',
        'wind_speed' : 'wind_speed',
        'wind_direction' : 'wind_direction',
        'rainfall_intensity' : 'rainfall_intensity',
        'accumulated_rainfall' : 'accumulated_rainfall'
        }
    
data_dict = {
    'time' : [],
    'speed' : [],
    'heading': [],
    'air_temperature': [],
    'humidity':[],
    'pressure':[],
    'wind_speed':[],
    'wind_direction':[],
    'rainfall_intensity':[],
    'accumulated_rainfall':[],
    'longitude':[],
    'latitude':[]
}
regex = r'POINT\((-?[0-9]+\.?[0-9]*) (-?[0-9]+\.?[0-9]*)\)'
epoch = datetime(2000,1,1)
for row in dr:
    date_time = parse(row['weather_sensor_date_time'])
    date_time = date_time.replace(tzinfo=None)
    dt = date_time - epoch
    timestamp = dt.total_seconds()
    data_dict['time'].append(timestamp)

    for field in data_dict.iterkeys():
        if field not in ('time', 'longitude', 'latitude'):
            data_dict[field].append(float(row[mapping[field]]))

    matches = re.match(regex, row['st_astext'])
    lng, lat = matches.groups()
    lng = float(lng)
    lat = float(lat)

    data_dict['longitude'].append(lng)
    data_dict['latitude'].append(lat)

f.close()
```

```python
nc.variables['time'][:] = np.array(data_dict['time'])
nc.variables['speed'][:] = np.array(data_dict['speed'])
nc.variables['heading'][:] = np.array(data_dict['heading'])
nc.variables['air_temperature'][:] = np.array(data_dict['air_temperature'])
nc.variables['humidity'][:] = np.array(data_dict['humidity'])
nc.variables['pressure'][:] = np.array(data_dict['pressure'])
nc.variables['wind_speed'][:] = np.array(data_dict['wind_speed'])
nc.variables['wind_direction'][:] = np.array(data_dict['wind_direction'])
nc.variables['heading'][:] = np.array(data_dict['heading'])
nc.variables['rainfall_intensity'][:] = np.array(data_dict['rainfall_intensity'])
nc.variables['accumulated_rainfall'][:] = np.array(data_dict['accumulated_rainfall'])
nc.variables['latitude'][:] = np.array(data_dict['latitude'])
nc.variables['longitude'][:] = np.array(data_dict['longitude'])
```

```python
nc.close()
```
