## Let's talk Rectangles!

I need a way of talking about rectangles and I need a way to describe a rectangle's relationship with other rectangles.
The notation and concepts applied here will be similar to the work I did with Intervals and Interval Trees.


A 2-d Rectangle is basically an interval extended into another (new) dimension.

We say $P$ is a rectangle and it has a lower bounds $(x_0, x_1)$ and an upper bounds $(x_0' x_1')$.

![Rectangle Def](https://docs.google.com/drawings/d/1G1FHZuD1tkem4mBxGKhruw0SnnDEo6RvYfWc7_4kmIw/pub?w=399&amp;h=350)

We define the property $x$ as the lower bounds and $x'$ as the upper bounds.
$$P(x) = (x_0, x_1)$$
$$P(x') = (x_0', x_1')$$

### Properties


#### Equivalency

We say a rectangle $P$ is equivalent to $Q$ if for all points $p$ in $P$, $p$ is also in $Q$ and for all points $q$ in
$Q$ that $q$ is also in $P$. This can be summarized as:

$$ P.x_n = Q.x_n \\
P.x_n' = Q.x_n'$$

#### Disjoint

![Disjoint](https://docs.google.com/drawings/d/1vKNbb6XLmvpdXEYWjrxabssVSxyQMSEAiEAJbleBx7g/pub?w=380&amp;h=291)

We say a rectangle $P$ is disjoint of rectangle $Q$ if for all points in $P$, there is no point $p \in P$ that is in
$Q$.

$P$ is disjoint from $Q$ if any of the following are met.

For any $n$ in the rectangle's dimensions:

 - If $P.x_n > Q.x_n'$ then $P$ is disjoint $Q$.
 - If $P.x_n' < Q.x_n$ then $P$ is disjoint $Q$.

I have a proof for the above, but it's quite intuitive so I won't post it here.

#### Contains
![Contains](https://docs.google.com/drawings/d/1-J-cLBDRtdIasZQRLS_JFcO-XuFowPV6g_QsZUAEfa0/pub?w=423&amp;h=463)

We say a rectangle $P$ contains $Q$ if for all points $q$ in $Q$, $q$ is also in $P$.

We can summarize the property looking at the bounds of the rectangle as follows:

If $P.x_n \le Q.x_n \le Q.x_n' \le P.x_n'$ then $P$ contains $Q$.

#### Intersects

![Intersects](https://docs.google.com/drawings/d/1Rkt5iX08Xew31hMwgYbEB-DZAalkr1FxCt2uDalRMCw/pub?w=477&amp;h=334)

We say a rectangle $P$ intersects $Q$ if there is at least one point $p$ in $P$ that is also in $Q$.

The easiest way to characterize this property is if $P$ is not disjoint $Q$.

#### Unions of Two Rectangles
![Union](https://docs.google.com/drawings/d/1gOYi72iBwDUFCVQ2QN68yVXqvHcT7UsGtFvF7f9Bvi8/pub?w=333&amp;h=235)

### The Code






```python
class Rectangle:
    ndim=None # Number of dimensions
    xmin=None # Tuple of minimum values
    xmax=None # Tuple of maximum values

    def __init__(self, xmin, xmax):
        '''
        xmin - a tuple of minimum values
        xmax - a tuple of maximum values

        Constructs a rectangle that is n-dimensional where n is the length of
        the tuple.  
        '''

        for x in zip(xmin, xmax):
            if not xmin <= xmax:
                raise ValueError("min tuple must be less than or equal to max tuple")
        self.ndim = len(xmin)

        self.xmin = xmin
        self.xmax = xmax


    def disjoint(self, other):
        if self.ndim != other.ndim:
            raise ValueError("rectangles aren't in the same dimension")
        for i in xrange(self.ndim):
            if self.xmin[i] > other.xmax[i]:
                return True
            if self.xmax[i] < other.xmin[i]:
                return True
        return False

    def contains(self, other):
        if self.ndim != other.ndim:
            raise ValueError("rectangles aren't in the same dimension")

        for i in xrange(self.ndim):
            if other.xmin[i] < self.xmin[i]:
                return False
            if other.xmax[i] > self.xmax[i]:
                return False
        return True

    def intersects(self, other):
        return not self.disjoint(other)
    
    def __repr__(self):
        return "Rectangle<%s,%s>" % (self.xmin, self.xmax)
```

```python
P = Rectangle((0,0), (1,1))
Q = Rectangle((0.9, 0.9), (1.2,1.2))

print P.intersects(Q)
print P.disjoint(Q)
print P.contains(Q)

Q = Rectangle((1,1), (1.5, 1.5))
print P.intersects(Q) # The corners touch
print Q.intersects(P)

Q = Rectangle((0.25, 0.25), (0.75, 0.75))
print P.contains(Q)
print P.intersects(Q)
```
