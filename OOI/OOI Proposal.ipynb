# OOI Bridge

These are some ideas we kicked around the office this morning. Each of these
are something that we can produce in a relatively short time frame with the
current team memebers from ASA. This is intended to be a listing of things that
we have at our finger tips and can produce and have already produced for
various projects. My idea here is not for the future of OOI but something that
can satisfy as a bridged solution in a short time frame, as the mature product
is being produced. These components are intended to be modular to the point
where different solutions can be swapped in and out with minimal effort.

Any combination of these would require a very brief investigative period to
determine proper cost and level of effort, this is intended mereley as a
catalog of technical skills and capabilities 

## Data and Data Access 

#### Members
 - Luke Campbell
 - Brian McKenna

### Capabilities and Scope

##### NetCDF Scraper off current Production

   Scrape the existing datasets off of OOI Production into formatted and
   CF-Compliant NetCDF datasets. These files then can be used as the primary
   data source for multiple catalog technologies as well as direct download
   support to Matlab, Python.

   The downside to this approach is that once production goes offline we'll
   need to identify a solution to continue to ingest data from OOI sensors and
   from the OMC data sources.

##### Direct Access and Scraping from OMC Datasets

   Identify a way to either use manufacturer drivers or use existing OOI
   instrument agent/driver code to parse OMC datasets to receive continuous
   data streams from OOI instruments. 

##### Risks
  
   We're looking at the driver and instrument agent code right now to identify
   the current state in this part of OOI and see what options we may have that
   aren't listed.
