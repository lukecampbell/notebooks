Then, at recovery 2014, the sensor was off by [1.08 0.30].

The output data product should blend these two together.

My proposed solution:
The default action is a linear interpolation between the two based on elapsed 
time - but we do not interpolate the coefficients (these could be a higher-
order polynomial). Rather, we interpolate between the two different outputs.

The computing steps are as follows:

1. We have data at time steps TIME.
2. We have a set of coefficients COEF1 at time CALTIME1.
3. We have another set of coefficients COEF2 at time CALTIME2.
4. For all TIME, the following holds: CALTIME1<=TIME<=CALTIME2
5. Compute a data set DAT1 based only on COEF1 as:
   DAT1 = polyval(COEF1,data);
6. Compute a data set DAT2 based only on COEF2 as:
   DAT2 = polyval(COEF2,data);
7. Compute weights to be applied to DAT1, such that the weights are
   1 at CALTIME1 decaying to 0 at CALTIME2:
   WEI1 = interp1([CALTIME1 CALTIME2],[1 0],TIME);
8. Ditto for DAT2, but increasing from 0 to 1:
   WEI2 = interp1([CALTIME1 CALTIME2],[0 1],TIME);
9. Note that the sum of WEI1 and WEI2, element-by-element, is a vector
   containing 1 everywhere.
10. Compute the output data product:
   DATAOUT = (WEI1.*DAT1) + (WEI2.*DAT2);

```python
sst = np.array([10.0, 10.2, 10.1, 9.8, 8.6]) # Temperature observations over 5 hours
time_steps = np.array([15, 16, 17, 18, 19],dtype=np.float32)

user_calibrations_post_deployment = np.array([1.03, 0.25]) # The sensor is off by 3% and a constant of 0.25 degrees
user_calibrations_post_recovery = np.array([1.06, 1]) # The sensor was recovered and was off by 1 degree

sst_corrected_post_deployment = np.polyval(user_calibrations_post_deployment, sst) # First set of polyval corrections
sst_corrected_post_recovery = np.polyval(user_calibrations_post_recovery, sst) # Second set of polyval corrections
# Interpolate between the two 
sst_corrected = np.empty_like(time_steps)
for i in xrange(len(time_steps)):
    w = 1.0 * i / (len(time_steps)-1)
    sst_corrected[i] = (1-w) * sst_corrected_post_deployment[i] + w * sst_corrected_post_recovery[i]

    
```

```python
# Plot it
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
fig = plt.subplot(111)
fig.set_title('Sea Surface Temperature')
fig.set_xlabel('Time (hour of day)')
fig.set_ylabel('Temperature (degrees in C)')
fig.plot(time_steps, sst, color='b', label='Observed Sea Surface Temperature')
fig.plot(time_steps, sst_corrected_post_deployment,'--', color='m', label='Post Deployment Corrected')
fig.plot(time_steps, sst_corrected_post_recovery, '--', color='g', label='Post Recovery Corrected')
fig.plot(time_steps, sst_corrected, color='r', label='Corrected Sea Surface Temperature')
fig.legend()
fig.scatter(time_steps, sst, color='b')
fig.scatter(time_steps, sst_corrected_post_deployment, color='m')
fig.scatter(time_steps, sst_corrected_post_recovery, color='g')
fig.scatter(time_steps, sst_corrected, color='r')
```

```python
# 4 different calibrations over time
cal0 = np.array([1.02, 1.0]) # 2% gain and .1 offset
cal1 = np.array([1.02, 2.0]) # .2 offset
cal2 = np.array([ 0.0003, 1.02, 3.0]) # 25% gain and no offset
cal3 = np.array([1.02, 4.0]) # wtf

# The times each calibration applies
cal_t = np.array([10, 20, 60, 80])



cals = np.array([cal0, cal1, cal2, cal3])

# Define the observation curve
t_i = [0, 25, 50, 75, 100]
x_i = [10, 12, 13, 12, 10]

p = np.polyfit(t_i, x_i, 2)

def f(x):
    return p[0] * x**2 + p[1] * x + p[2]

# Actual observation vectors
# t is the time
t = np.arange(0,100)
# x is the observation values (unitless)
x = f(t)
```

```python
# Calibration

polyvals = np.zeros_like(x)
x0s = np.zeros_like(x)
x1s = np.zeros_like(x)
a = np.searchsorted(t, cal_t)
print a
j = 0
section_a = 0
for i, x_i in enumerate(x):
    if (j+1) >= len(a):
        polyvals[i] = x_i
        section_a += 1
        continue
    elif i > a[j+1]:
        j = min(j + 1, len(a)-1)
        print j
        if (j+1 < len(a)) and (a[j] <= i and i <= a[j+1]): 
            x0 = np.polyval(cals[j], x_i)
            print "x0=", x0
            x0s[i] = x0
            x1 = np.polyval(cals[j+1], x_i)
            x1s[i] = x1
            w = (t[i] - t[a[j]] * 1.0) / ( t[a[j+1]] - t[a[j]] * 1.0)

            polyvals[i] = x0 * (1 - w) + x1 * w
        else:
            polyvals[i] = x_i
    elif a[j] <= i and i <= a[j+1]: 
        x0 = np.polyval(cals[j], x_i)
        x0s[i] = x0
        x1 = np.polyval(cals[j+1], x_i)
        x1s[i] = x1
        w = (t[i] - t[a[j]] * 1.0) / ( t[a[j+1]] - t[a[j]] * 1.0)
        polyvals[i] = x0 * (1 - w) + x1 * w
    else:
        polyvals[i] = x_i
    
print "a_count:",section_a
```

```python
# Plot it
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
fig = plt.subplot(111)
fig.set_title('Observations')
fig.set_xlabel('Time')
fig.set_ylabel('Data')
fig.plot(t, x, color='b', label='Observations')
fig.plot(t, polyvals, color='r', label='Polyvals')
fig.plot(t, x0s, color='g', label='X0')
fig.plot(t, x1s, color='m', label='X1')
fig.legend()
#fig.scatter(t, x, color='b')
#fig.scatter(t, polyvals, color='r')
```
