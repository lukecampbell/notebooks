```python
# First step is probably to extract the zipfile

filepath = '/tmp/lrt/Data_QC_Lookup_Table_Local_Range_Test.lrt'

from zipfile import ZipFile
from csv import DictReader, reader
import numpy as np
from uuid import uuid4
import os

def guid():
    return uuid4().hex

object_store = {}



def main():
    global object_store
    with ZipFile(filepath) as zp:
        files = zp.namelist()
        table_map = {}
        for f in files:
            if os.path.basename(f) != 'master.csv':
                with zp.open(f) as table:
                    table_map[f] = set_table(table)
        with zp.open('master.csv') as master:
            set_master(master, table_map)


def set_table(table):
    global object_store
    headers = reader(table).next()
    data = np.genfromtxt(table, delimiter=',')

    record = {}
    for i,h in enumerate(headers):
        record[h] = data[:, i]

    key = guid()

    object_store[key] = record
    return key


def set_master(master, table_map):
    global object_store
    dr = DictReader(master)
    for row in dr:
        rd = row['reference_designator']
        dp = row['data_product_name']
        units = row['units']
        author = row['author']
        date = row['date_updated']
        table_key = row['table']

        inner_record = {
            "units":units,
            "author":author,
            "date":date,
            "table_key" : table_map[table_key]
        }

        if rd in object_store:
            entry = object_store[rd]
            if dp not in entry:
                entry[dp] = [ inner_record ]
            else:
                entry[dp].append(inner_record)
            object_store[rd] = entry

        else:
            object_store[rd] = {
                dp : [ inner_record ]
            }
main()
```
