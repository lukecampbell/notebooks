# Interpolating Functions

We have a problem where we need to be able to aggregate non-contiguous and non-continuous datasets and apply functions to the discrete data that is available in the data. For example, consider the OPTATTN data product:

> OPTATTN is an L2 product in that computation requires the raw signals emanating from a 
> properly calibrated and configured instrument as well as water temperature (TEMPWAT) and 
> salinity (PRACSAL) derived from a co-located and synchronized CTD.

[OPTATTN Data Product Specification](https://alfresco.oceanobservatories.org/alfresco/d/d/workspace/SpacesStore/28089882-716e-4fa8-b4b6-c703f5630882/1341-00690_Data_Product_SPEC_OPTATTN_OOI.pdf)

We don't have a low-level way to synchronize the two instruments so that the sampling and timestamps are synchronized between them which leaves us with two distinct discrete functions. One way we can artificially synchronize the two discrete functions is by using a linear basis function. We can interpolate any value of the function between two discrete points by using a set of basis functions, which allows us to determine points along a function with a different domain, specifically the discrete domain of another instrument.

Consider the function $u(t_i)$ which represents the temperature measured by a CTD at a specific time $t_i$, where $t$ is a continuous variable and $t_i$ is a discrete variable. We can identify a function $u_i(t)$ that is defined as

$$ u(t) = u_i(t_i) \cdot (1-t) + u_i(t_{i+1}) \cdot t $$

Now that we have a continuous function between $t_0$ and $t_n$ we can synchronize the two datasets. 

# Example

Consider that we have a colocated instrument that measures temperature at an irregular sampling. We also have a local instrument that measures temperature as well and we want to present a data product that represents the mean temperature between these two instruments.

We can calculate the mean between overlapping discrete points between the two datasets but there are none. There are no exactly identical times where we can reliably determine the two values since the domains don't intersect. Using an interpolation for the colocated instrument we can construct a function that is continuous in the domain of the local instrument, then determine the mean values for the data product.

```python
from scipy.interpolate import interp1d
# Define the continuous function over the colocated instrument's domain
f = interp1d(colocated_time, colocated_temp)

# Map the continous function to the domain of the local instrument
applicable_domain_idx = np.where((instrument_time >= colocated_time[0]) & (instrument_time <= colocated_time[-1]))[0]
applicable_domain = instrument_time[applicable_domain_idx]

u = f(applicable_domain)
v = instrument_temp[applicable_domain_idx]
```

```python
fig = plt.subplot(111)
fig.plot(applicable_domain, u, label='Collocated Temperature Interpolated')
fig.plot(applicable_domain, v, color='r', label='Local Temperature')
fig.set_title('Local vs Colocated Temperature Interpolated')
fig.set_xlabel('Time (s)')
fig.set_ylabel('Temperature (deg_F)')
fig.legend()
```

Now each point lies in the same domain, we can calculate our data product!

```python
data_product = np.mean([u,v], axis=0)
```

```python
fig = plt.subplot(111)
fig.plot(applicable_domain, u, label='Collocated Temperature Interpolated')
fig.plot(applicable_domain, v, color='r', label='Local Temperature')
fig.plot(applicable_domain, data_product, color='g', label='Mean Temperature')
fig.set_title('Local vs Colocated Temperature Interpolated')
fig.set_xlabel('Time (s)')
fig.set_ylabel('Temperature (deg_F)')
fig.legend()
```

# Cosine interpolation
Linear interpolation results in discontinuities at each point. Often a smoother interpolating function is desirable, perhaps the simplest is cosine interpolation. A suitable orientated piece of a cosine function serves to provide a smooth transition between adjacent segments.

```python
def cosine_interp(x_i, y_i, x):
    x_new = np.searchsorted(x_i, x)
    x_new = x_new.clip(1, len(x_i)-1).astype(int)
    x_low = x_new - 1
    ix = (x - x_i[x_low]) / (x_i[x_new] - x_i[x_low])
    mu = (1 - np.cos(ix * np.pi)) / 2
    y = (y_i[x_low] * (1 - mu) + y_i[x_new] * mu)
    return y

x_i = np.array([1.2, 2.3,   3.1, 4.22, 5.61, 6, 7.53, 8.73])
y_i = np.array([3.5, 1.24, 1.69,  1.1,  3.3, 0, 2.12, 0.75])

x = np.linspace(x_i[0], x_i[-1], 200)
y = cosine_interp(x_i, y_i, x)
```
